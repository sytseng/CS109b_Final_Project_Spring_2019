## Assumptions about the data

All the mouse data, which has already been filtered for you, is split up into two different sets. Mouse set 1 corresponds to mice `'2'`, `'3'`, `'4'`, and `'5'` and mouse set 2 corresponds to mice `'6'`, `'7'`, `'8'`, `'9'`, and `'10'`. __Don't__ assume that all the OTU's in the first mouse set are in the second mouse set, and vice versa. Additionally, __don't__ assume that all of the days in mouse set 1 are the same in mouse set 2. However, __every mouse within the same set all have the same OTU's and days__. For example, if mouse `'2'` has data for OTU `'OTU_91'`, then we can assume that mice `'3'`, `'4'`, and `'5'` also have data fir `'OTU_91'`. But the mice in mouse set 2 do not necessarily have to have data for `'OTU_91'`.

## Structure of the mouse set objects

The object `mouse_set_1` that you just loaded is a _dict_ object with four different entries:
1. `'reads'`
   * This key maps to the raw count data of each mouse that was outputted by [DADA2](https://github.com/benjjneb/dada2).


2. `'qpcr'`
   * This key maps to the qpcr (total abundance) data for each day for each mouse.

4. `'otu_taxonomy'`
   * This is a table that tells you the (approximate) taxonomy for each of the OTUs in the list as well as the 16S rRNA sequence that was used to barcode the OTU.
   
5. `'times'`
   * This is a _list_ object that converts the string time into a float time. More explanation will be below.
   

```python
reads = mouse_set_1['reads']
qpcr = mouse_set_1['qpcr']
otu_taxonomy = mouse_set_1['otu_taxonomy']
times = mouse_set_1['times']

reads2 = mouse_set_2['reads']
qpcr2 = mouse_set_2['qpcr']
otu_taxonomy2 = mouse_set_2['otu_taxonomy']
times2 = mouse_set_2['times']
```

```python
modelString = """
model{
    
    # dirichlet process
    for (clus in 1:C){alpha[clus] ~ dgamma(1,1)}
    pi[1:C] ~ ddirch(alpha[1:C])
    
    for (i in 1:N){
        c[i] ~ dcat(pi[1:C]);
    }
    
    
    # interaction
    sig_b_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_b = sqrt(1/sig_b_2_inv);
    
    for (i in 1:C){
        for (j in 1:C){
            this_b[i,j] ~ dnorm(0, sig_b_2_inv);
            b[i,j] = ifelse(i==j, 0, this_b[i,j]);
        }
    }
    
    
    # self interaction
    sig_a_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_a = sqrt(1/sig_a_2_inv);
    
    for (i in 1:N){
        a1[i] ~ dnorm(0,sig_a_2_inv);
        a2[i] ~ dnorm(0,sig_a_2_inv);
    }
    
    # edge selection
    pi_z ~ dunif(0,1)
    for (i in 1:C){
        for (j in 1:C){
            z[i,j] ~ dbern(pi_z);
        }
    }
    
    # dynamics
    sig_w_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_w = sqrt(1/sig_w_2_inv);
    
    # initialize x[k=1, i], and bzx[k=1,i,j]
    for (i in 1:N){
        x[1,i] ~ dnorm(0, sig_w_2_inv/times[1]);
    }
    
    for (i in 1:N){
        for (j in 1:N){
            bzx[1,i,j] = 0;
        }
    }
    
    # perturbation
    for (i in 1:N){
        P_fat[i] ~ dnorm(0,0.0001);
        P_pos[i] ~ dnorm(0,0.0001);
        P_neg[i] ~ dnorm(0,0.0001);
    }
    
    
    for (k in (2:K)){
        for (i in 1:N){
            for (j in 1:N){
                bzx[k,i,j] = b[c[i],c[j]]*z[c[i],c[j]]*x[k-1,j];
            }
            x[k,i] ~ dnorm(x[k-1,i] + x[k-1,i]*(a1[i] + a2[i]*x[k-1,i] + sum(bzx[k,i,]) + fat[k-1]*P_fat[i] + pos[k-1]*P_pos[i] + neg[k-1]*P_neg[i]), sig_w_2_inv/(times[k]-times[k-1]));
        }
    }
    
    # constrain and measurement model
    sig_q_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_q = sqrt(1/sig_q_2_inv);
    
    for (k in 1:K){
        for (i in 1:N){
            q[k,i] ~ dnorm(x[k,i], sig_q_2_inv);        
        }
    }

    for (k in 1:K){
        for (i in 1:N){
            phi[k, i] = r[k]*(q[k,i]/sum(q[k, ]));
            eps[k, i] = a0/(q[k,i]/sum(q[k, ])) + a_1;
        } 
    }

    for (k in 1:K){
        for (i in 1:N){
            y[k,i] ~ dnegbin(phi[k,i], eps[k,i]);
        }
    }

    for (k in 1:K){
        Q[k] ~ dnorm(sum(q[k, ]), 1/pow(sigma_Q,2));
    }
}"""
```

```python
modelString = """
model{        
    # interaction
    sig_b_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_b = sqrt(1/sig_b_2_inv);
    
    for (i in 1:N){
        for (j in 1:N){
            this_b[i,j] ~ dnorm(0, sig_b_2_inv);
            b[i,j] = ifelse(i==j, 0, this_b[i,j]);
        }
    }
    
    
    # self interaction
    sig_a_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_a = sqrt(1/sig_a_2_inv);
    
    for (i in 1:N){
        a1[i] ~ dnorm(0,sig_a_2_inv);
        a2[i] ~ dnorm(0,sig_a_2_inv);
    }
    
    # edge selection
    pi_z ~ dunif(0,1)
    for (i in 1:N){
        for (j in 1:N){
            z[i,j] ~ dbern(pi_z);
        }
    }
    
    # dynamics
    sig_w_2_inv ~ dgamma(0.1,0.1);   # I don't know what prior we should use
    sig_w = sqrt(1/sig_w_2_inv);
    
    # initialize x[k=1, i], and bzx[k=1,i,j]
    for (i in 1:N){
        x[1,i] ~ dnorm(0, sig_w_2_inv);
    }
    
    for (i in 1:N){
        for (j in 1:N){
            bzx[1,i,j] = 0;
        }
    }
    
    # perturbation
    for (i in 1:N){
        P_fat[i] ~ dnorm(0,0.0001);
        P_pos[i] ~ dnorm(0,0.0001);
        P_neg[i] ~ dnorm(0,0.0001);
    }
    
    
    for (k in (2:K)){
        for (i in 1:N){
            for (j in 1:N){
                bzx[k,i,j] = b[i,j]*z[i,j]*x[k-1,j];
            }
            x[k,i] ~ dnorm(x[k-1,i] + x[k-1,i]*(a1[i] + a2[i]*x[k-1,i] + sum(bzx[k,i,]) + fat[k-1]*P_fat[i] + pos[k-1]*P_pos[i] + neg[k-1]*P_neg[i]), sig_w_2_inv/(times[k]-times[k-1]));
        }
    }

}"""
```

```python
def plot_trace(samples, varname, entry=0):
    plt.plot()
    sample_array = samples[varname]

    vec_len, num_samples, num_chains = sample_array.shape
    for cur_chain in range(num_chains):
        cur_label = "Chain {}".format(cur_chain)
        plt.plot(range(num_samples),sample_array[entry,:,cur_chain], label=cur_label, linewidth = 0.5)
    plt.legend()
    plt.show()
```
